```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype='int')
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split #train and test set split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
from sklearn.linear_model import LinearRegression #importing linear regression function in sklearn library
from sklearn.model_selection import GridSearchCV
param_grid={'copy_X': [True,False], 
               'fit_intercept': [True,False], 
               'n_jobs': [1,5,10,15,None], 
               'positive': [True,False]}
grid=GridSearchCV(LinearRegression(),param_grid,verbose=3,n_jobs=-1,refit=True)
```

```python
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
print("The r2_score value for the best parameters {}".format(grid.best_params_))
```

```python
y_pred=grid.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
